## Merging Dataframes

```python
import pandas as pd
```

```python
staff_df = pd.DataFrame([{'Name':'Kelly','Role':'Director of HR'},
                        {'Name':'Sally','Role':'Course liasion'},
                        {'Name':'James','Role':'Grader'}])
staff_df = staff_df.set_index('Name')

student_df = pd.DataFrame([{'Name':'James','School':'Business'},
                        {'Name':'Mike','School':'Law'},
                        {'Name':'Sally','School':'Engineering'}])
student_df = student_df.set_index('Name')

print(student_df.head())
print(staff_df.head())
```

```python
# outer join is union
pd.merge(staff_df,student_df,how='outer',left_index=True,right_index=True)
```

```python
# inner join is intersect
pd.merge(staff_df,student_df,how='inner',left_index=True,right_index=True)
```

```python
# list all components in the left table with addition info from right table
pd.merge(staff_df,student_df,how='left',left_index=True,right_index=True)
```

```python
pd.merge(staff_df,student_df,how='right',left_index=True,right_index=True)
```

```python
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()
pd.merge(staff_df,student_df,how='right',on = 'Name')
```

### Method chaining (pandorable)

```python
(df.where(df['SUMLEV'] == 50)
     .dropna()
     .set_index(['STNAME','CTYNAME'])
     .rename(columns = {'ESTIMATESBASE2010':'Estimates Base 2010'}))
```

```python
df = df[df['SUMLEV']==50]
df.set_index(['STNAME','CTYNAME'],inplace = True)
df.rename(columns = {'ESTIMATESBASE2010':'Estimates Base 2010'})
```

```python
def first_approach():
    global df
    return (df.where(df['SUMLEV'] == 50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns = {'ESTIMATESBASE2010':'Estimates Base 2010'}))

df = pd.read_csv('datasets/census.csv')

timeit.timeit(first_approach,number=10)
```

```python
def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'],inplace = True)
    return new_df.rename(columns = {'ESTIMATESBASE2010':'Estimates Base 2010'})
df = pd.read_csv('datasets/census.csv')

timeit.timeit(second_approach,number=10)
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['min'] = np.min(data)
    row['max'] = np.max(data)
    return row
```

## Group by
* In pandas, this is refered to as the split-apply-combine pattern

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
%%timeit -n 3

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
#     print('Counties in state ' + state + ' have Population ' + str(avg))
```

```python
%%timeit -n 3
# first return value is the key we want to group for, the second one is the dataframe 
for group,frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
#     print('Counties in state ' + group + ' have Population ' + str(avg))
    
```

### Customize functions with `groupby()`

```python
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

# if no column name pass to groupby function, it would use index as default
for group, frame in df.groupby(set_batch_number):
    print('There are '+ str(len(frame)) + ' records in group ' + str(group) + ' for processing')
```

### Dealing with muti-indexes

```python
df = pd.read_csv('datasets/listings.csv')
df.head()
```

```python
df = df.set_index(['cancellation_policy','review_scores_value'])

for group, frame in df.groupby(level = (0,1)):
    print(group)
```

```python
def grouping_func(item):
    if item[1] == 10.0:
        return(item[0],"10.0")
    else:
        return(item[0],"not 10.0")
    
for group,frame in df.groupby(by = grouping_func):
    print(group)
```

```python
df.head()
```

### Aggregation

```python
df = df.reset_index()
df.groupby('cancellation_policy').agg({"review_scores_value":np.average})
```

```python
df.groupby('cancellation_policy').agg({"review_scores_value":np.nanmean})
```

```python
df.groupby('cancellation_policy').agg({"review_scores_value":(np.nanmean,np.nanstd),
                                       'reviews_per_month':np.nanmean})
```

### Transformation
* `agg()` returns a single value per column
* `transform()` returns an object that is the same size as the group , so we could merge the result from it to original table

```python
cols = ['cancellation_policy','review_scores_value']
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'},axis='columns',inplace=True)
df = df.merge(transform_df, left_index=True,right_index=True)
df.head()
```

```python
df['mean_diff'] = np.absolute(df['review_scores_value'] - df['mean_review_scores'])
df['mean_diff'].head()
```

### Filtering

```python
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

### Applying

```python
df = pd.read_csv('datasets/listings.csv')
df = df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
def calc_mean_review_scores(group):
    # group is an dataframe here
    avg = np.nanmean(group['review_scores_value'])
    group['review_scores_mean'] = np.abs(avg - group['review_scores_value'])
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

## Pivot Table

```python
df = pd.read_csv('datasets/cwurData.csv')
df.head(10)
```

```python
def my_rank(national_rank):
    if national_rank <= 100:
        return 'First Tier Top University'
    if national_rank <= 200:
        return "Second Tier Top University"
    if national_rank <= 300:
        return "Third Tier Top University"
    return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply(my_rank)
df['Rank_Level'].head(10)
```

```python
df.head()
```

```python
df.pivot_table(values='score',index = 'country',columns = 'Rank_Level',aggfunc=np.nanmean)
```

```python
df.pivot_table(values='score',index = 'country',columns = 'Rank_Level',aggfunc=[np.nanmean,np.max]).head()
```

```python
df.pivot_table(values='score',index = 'country',columns = 'Rank_Level',aggfunc=np.nanmean,margins=True)
```

```python
new_df = df.pivot_table(values='score',index = 'country',columns = 'Rank_Level',aggfunc=[np.nanmean,np.max])
new_df['nanmean']['First Tier Top University'].idxmax()
```

```python
new_df = new_df.stack()
new_df.head(10)
```
